```python
dataset=pd.read_csv("Nat_Gas.csv")
```

```python
dataset
```

```python
# Convert to DataFrame
df = pd.DataFrame(dataset)
df["Dates"] = pd.to_datetime(df["Dates"])
df["Days"] = (df["Dates"] - df["Dates"].min()).dt.days  # Convert dates to numerical values
```

```python
df["Dates"]
```

```python
# Train a linear regression model
X = df[["Days"]]
y = df["Prices"]
model = LinearRegression()
model.fit(X, y)
```

```python
def predict_gas_price(date_str):
    date = datetime.strptime(date_str, "%m/%d/%y")
    days_since_start = (date - df["Dates"].min()).days
    predicted_price = model.predict([[days_since_start]])[0]
    return round(predicted_price, 2)
```

```python
# Predict prices for the next year
future_dates = [df["Dates"].max() + timedelta(days=30*i) for i in range(1, 13)]
future_days = [(date - df["Dates"].min()).days for date in future_dates]
future_prices = model.predict(np.array(future_days).reshape(-1, 1))
```

```python
# Example usage
print(predict_gas_price("10/31/25"))
```
