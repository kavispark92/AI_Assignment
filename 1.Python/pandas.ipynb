```python
df['Total']=pd.eval("df.ssc_p+df.hsc_p+df.degree_p+df.etest_p+df.mba_p")
display(df)
```

```python
pd.eval("Avg=df.Total/5",target=df,inplace=True)
pd.eval("per=(df.Total/500)*100",target=df,inplace=True)
display(df)
```
