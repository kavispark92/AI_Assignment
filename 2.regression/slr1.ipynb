```python
#pandas is a python library. it contains dataset to create data analysis workflows
import pandas as pd
```

```python
#We are loading the salary-_data in the dataset variable which is in the pandas read_csv file
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#we can see what is inside the dataset
dataset
```

```python
#assigning inde as variable to store input data
inde=dataset[["YearsExperience"]]
```

```python
inde
```

```python
##assigning dep as variable to store output data
dep=dataset[["Salary"]]
```

```python
dep
```

```python
#importing train, test, split procedure from sklearn model selection
#assigning input,output parameters to resp variables,test size should be less than train size.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inde,dep,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_train
```

```python
X_test
```

```python
#creating a model by importing linear regression from machine learning
#creating a regressor variable to assign the Linear regresson class
#by using fit method we are passing training set as input
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#coef_method is used to calculate the weight of the model
weight=regressor.coef_
```

```python
weight
```

```python
#intercept method is used to calculate the bias of the model
bias=regressor.intercept_
```

```python
bias
```

```python
#creating the variable Y_predict to predict the avlue of X_test
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
#findinf R_score by using the parameter Y_test,Y_pred
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#save the model(using pickle python module) with new name,Wb-write binary,rb-read binary
import pickle
filename="linear_final_model.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loading the model to get predictedoutput
load_model=pickle.load(open("linear_final_model.sav",'rb'))
result=load_model.predict([[15]])
```
