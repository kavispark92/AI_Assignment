```python
raw_df['PRICE'].value_counts()
```

```python
raw_df['VOTES'].replace(to_replace=['-','NEW','Opening'],value='0',inplace=True)
```

```python
raw_df['PRICE'] = raw_df['PRICE'].astype('int64') 
raw_df['RATING'] = raw_df['RATING'].astype('float64') 
raw_df['VOTES'] = raw_df['VOTES'].astype('int64') 
```

```python
raw_df['TIMING'].value_counts()
```

```python
temp_df=raw_df['TIMING'].str.split("(", n=1, expand=True)
temp_df
```

```python
raw_df['TIMING']=temp_df[0]
raw_df['Days_Open']=temp_df[1]
raw_df
```

```python
raw_df['Days_Open'] = raw_df['Days_Open'].str.replace('\)','',regex=True) 
raw_df.head()
```

```python
useless_df=(raw_df['RATING']==0 )|( raw_df["VOTES"]==0)
```

```python
raw_df=raw_df[~useless_df]
```

```python
raw_df
```

```python
raw_df["RATING_TYPE"].value_counts()
```

```python
raw_df['RATING_TYPE'].replace(to_replace='Excelente' , value='Excellent', inplace=True) 
raw_df['RATING_TYPE'].replace(to_replace=['Veľmi dobré','Bardzo dobrze','Muy Bueno','Velmi dobré'] , value='Very Good',inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Skvělá volba','Dobrze','Bueno','Buono','Dobré','Bom','Skvělé'] , value='Good',inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Priemer','Média','Çok iyi'] , value='Average', inplace=True) 
raw_df['RATING_TYPE'].replace(to_replace=['Průměr','Promedio','Ortalama','Muito Bom','İyi'] , value='Poor', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Baik','Biasa','Media','Sangat Baik'] , value='Very Poor', inplace=True) 
```

```python
raw_df["RATING_TYPE"].value_counts()
```

```python
raw_df["REGION"].value_counts()
```

```python
raw_df["REGION"]=raw_df["REGION"].str.replace('[a-zA-Z].+--','',regex=True)
```

```python
raw_df["REGION"]=raw_df["REGION"].str.replace('west|West|east|East','',regex=True)
```

```python
#3)Which is best food truck in mumbai?
foodtruck_df =zomato_df[zomato_df["CUSINE TYPE"]=="Food Truck"]
foodtruck_df.sort_values(by='RATING',ascending=False).head(3)
```

```python
# 4) Which are the Top 10 highest rated Seafood Restaurant in Mumbai?
seafood_df =zomato_df[zomato_df["CUSINE_CATEGORY"]=="Seafood"]

seafood_df.sort_values(by='RATING',ascending=False).head(10)
```

```python
#5)  Which places have the highest rated restaurant for each Cuisine Type in Mumbai
highest_df=zomato_df[zomato_df['RATING']>=4.5]
highest_df
highest_rated_price_df = highest_df.groupby(by=['REGION', 'CUSINE TYPE'])['PRICE'].mean().reset_index() 
highest_rated_price_df.head() 
```

```python
#6) Which areas have a large number of Chinese Restaurant Market
chinese_df=zomato_df[zomato_df["CUSINE_CATEGORY"]=="Chinese"]
chinese_df.sort_values(by='REGION',ascending=True,inplace=True)
chinese_df
```

```python
large_chi_df = chinese_df.groupby("REGION")["CUSINE_CATEGORY"].nunique().reset_index()
large_chi_df
```

```python
large_chi_df = chinese_df.groupby("REGION").agg({'RATING_TYPE' : 'nunique', 'NAME' : 'count','PRICE':'mean'}).reset_index()
large_chi_df.sort_values(by='PRICE',ascending=False,inplace=True)
large_chi_df
```

```python
chinese_df
```

```python
large_chi_df = chinese_df.groupby(['REGION','CUSINE TYPE']).agg({'PRICE' : 'mean'}).reset_index()
large_chi_df.sort_values(by='REGION',ascending=False,inplace=True)
large_chi_df
```
